### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math as m

class Circle:
    def __init__(self,radius):
        self.radius = radius

    def Area(self):
        print("Area of Circle:",m.pi * self.radius ** 2)
    def Perimeter(self):
        print("Perimeter of Circle:",2 * m.pi * self.radius)
        
c1 = Circle(5)
c1.Area()
c1.Perimeter()
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank_Account:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance

    def Deposit(self,amount):
        if amount < 0:
             print("Please Enter Valid Amount to Deposit")
        else:
               self.balance += amount
            
    def Withdraw(self,amount):
        if amount < 0:
            print("Please Enter Valid Amount to Withdraw")
        elif amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance -= amount
            
    def Check_Balance(self):
        print("Your Current Account Balance:",self.balance)

name = input("Enter Your Name:")
balance = float(input("Enter your initial balance"))

b1 = Bank_Account(name,balance)
b1.Check_Balance()

deposit_Amount  = float(input("Enter the amount to Deposit:"))
b1.Deposit(deposit_Amount)
b1.Check_Balance()

withdraw_Amount = float(input("Enter the amount to Withdraw:"))
b1.Withdraw(withdraw_Amount)
b1.Check_Balance()
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self):
        self.item_name = None
        self.price = 0
        self.quantity = 0

    def add_item(self, item_name, price, quantity):
        self.item_name = item_name
        self.price = price
        self.quantity = quantity
        print(f"Item {item_name} added with price: ${price} and quantity: {quantity}.")


    def remove_item(self):
        if self.item_name is not None:
            print(f"Item {self.item_name} removed from inventory.")
            self.item_name = None
            self.price = 0
            self.quantity = 0
        else:
            print("No item to remove. The inventory is empty.")

    
    def update_item(self, price=None, quantity=None):
        if self.item_name is not None:
            if price is not None:
                self.price = price
                print(f"Updated price of {self.item_name} to ${price}.")
            if quantity is not None:
                self.quantity = quantity
                print(f"Updated quantity of {self.item_name} to {quantity}.")
        else:
            print("No item found in the inventory to update.")
            
    def display_inventory(self):
        if self.item_name is not None:
            print("\nCurrent Inventory:")
            print(f"Item: {self.item_name}, Price: ${self.price}, Quantity: {self.quantity}")
        else:
            print("Inventory is empty.")


   
inventory = Inventory()

inventory.add_item("Laptop", 1000, 5)

inventory.display_inventory()

inventory.update_item(price=1200, quantity=10)

inventory.display_inventory()

inventory.remove_item()

inventory.display_inventory()
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour, minute):
        self.hour = hour
        self.minute = minute

    def add_time(self, other):
        total_minutes = self.minute + other.minute
        extra_hours = total_minutes // 60
        total_hours = self.hour + other.hour + extra_hours
        final_minutes = total_minutes % 60
        return Time(total_hours, final_minutes)

    def display(self):
        print(f"Time: {self.hour} hours {self.minute} minutes")

time1 = Time(2, 45)
time2 = Time(1, 30)

result = time1.add_time(time2)
result.display()
```
