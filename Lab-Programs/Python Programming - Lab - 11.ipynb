### 01) WAP to create Calculator module which defines functions like add, sub,mul and div. 
### Create another .py file that uses the functions available in Calculator module.

```python
print("Choose an operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter your choice (1/2/3/4): "))
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

match choice:
    case 1:
        print("Addition:",add(a, b))
    case 2:
        print("Subtraction:", sub(a, b))
    case 3:
        print("Multiplication:",mul(a, b))
    case 4:
        print("Division:",div(a, b))
    case _:
        print("Invalid choice! Please select a valid operation.")
```

### 06) WAP to generate 100 random lottery tickets and pick two lucky tickets from it and announce them as Winner and Runner up respectively.

```python
tickets = []
while (len(tickets) < 100):
    num = random.randint(1,1000)
    if num not in tickets:
        tickets.append(num)
print('Winner Ticket: ',tickets[0])
print('Runner-up Ticket:',tickets[1])
```

### 11) WAP to demonstrate the use of date time module.

```python
d = date(2025,2,10)
print('Date:',d)

t = time(10,40,30)
print('time:',t)

dt = datetime.now()
print('Date & Time',dt)

# string type date to datetime object -> strptime()

date_s = '2025-2-10 14:30:59'
df = datetime.strptime(date_s,"%Y-%m-%d  %H:%M:%S")
print(type(df))
print(df)

# datetime object to string type -> strftime

s = datetime.strftime(dt,"%d,%B-%Y %H:%M:%S")
print(s)
print(type(s))
```
