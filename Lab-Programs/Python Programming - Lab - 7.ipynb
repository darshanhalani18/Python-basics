### 01) WAP to iterate over a set.

### 09) WAP to check whether two given strings are anagram or not using set.

```python
s1 = set(input('Enter String-1:'))
s2 = set(input('Enter String-2:'))

if len(s1) == len(s2) and s1 == s2:
    print('Strings are Anagram')
else:
    print('Strings are not Anagram')
```

### 11) WAP to count number of vowels in given string using set.

```python
s = input("Enter a String:")
vowels = {'a','e','i','o','u'}
count = 0

for char in s.lower():
    if char in vowels:
        count += 1
print(count)
```

### 12) WAP to check if a given string is binary string or not.

```python
s = input('Enter a String: ')
binaryChar = {'0','1'}
isValid = True

for char in s:
    if char not in binaryChar:
        isValid = False
        break
if isValid:
    print('The string is a binary string')
else:
    print('The string is not a binary string.')
```

### 14) WAP to find the sum of all items (values) in a dictionary given by user. (Assume: values are numeric)

```python
n = int(input('Enter the number of items in dictionary:'))
d = {}

for i in range(n):
    key = input('Enter key:')
    value = int(input(f'Enter value of {key}:'))
    d[key] = value

s = sum(d.values())
print('Sum of all items:',s)
```

### 15) WAP to handle missing keys in dictionaries.
#### Example : Given, dict1 = {'a': 5, 'c': 8, 'e': 2} 
#### if you look for key = 'd', the message given should be 'Key Not Found', otherwise print the value of 'd' in dict1.

```python
dict = {'a':5,'c':8,'e':2}
key = input('Enter the key you want to print the value of:')
if key in list(dict.keys()):
    print(dict[key])
else:
    print('Key not found')
```
