### 13) WAP to find second largest number in a list.

```python
li = [12,23,20,50,100]
max=li[0]
secondMax=li[0]

for i in li:
    if i > max:
        secondMax = max
        max = i
print('Second largest element:',secondMax)
```

### 14) WAP to extract elements with frequency greater than K.

```python
k=int(input("Enter Frequency K: "))
temp=[]
l=[10,20,30,40,20,20,30,40,20,30]
for i in l:
    if l.count(i)>k and temp.count(i)==0:
        temp.append(i)
temp
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
li = input('Enter list of fruits:').split(',')
ans = [i for i in li if i[0]=='b']
print(ans)
```

### 17) WAP to create a list of common elements from given two lists.

```python
li= [10,20,30,40,50,60]
li = [20,40,60,80]
ans=[]
for i in li:
    if li.count(i)>0 and ans.count(i)==0:
        ans.append(i)
print(ans)
```
